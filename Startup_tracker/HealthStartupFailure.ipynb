```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd

# Setup headless Chrome
options = Options()
options.add_argument("--headless")
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get("https://www.cbinsights.com/research/startup-failure-post-mortem/")

# Get the HTML after JavaScript loads
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

data = []
for h2 in soup.find_all("h2"):
    name = h2.get_text(strip=True)
    next_p = h2.find_next("p")
    if next_p:
        reason = next_p.get_text(strip=True)
        if "health" in reason.lower():
            data.append({
                "Startup Name": name,
                "Sector": "Digital Health",
                "Year": "Unknown",
                "Reason for Failure": reason,
                "Source URL": driver.current_url
            })

driver.quit()

df = pd.DataFrame(data)
df.to_csv("digital_health_failures.csv", index=False)
df.head()
```
